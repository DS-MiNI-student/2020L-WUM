```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./bank.csv", sep=";")
```

```python
data.head()
```

```python
data.describe()
```

```python
# Nie ma braków danych
```

```python
data.isnull().sum().sort_index()
```

```python
# Najwięcej jest trzydziestolatków
```

```python
plt.figure(figsize=(10,6))
sns.distplot(data.age, label='Skewness: %.2f' % (data.age.skew()))
plt.legend()
plt.show()
```

```python
# Brak widocznych zależności między wiekiem i pozostałymi zmiennymi
```

```python
sns.pairplot(data)
```

```python
# Widaź kilka obserwacji odstających - max wiek to 87, co nie wydaje się podejrzane
```

```python
sns.boxplot(data.age)
```

```python
# Najmłodsi są single, najstarsi rozwiedzeni
```

```python
sns.catplot(y="age", x="marital", kind = "box", data=data)
```

```python
# Najstarsi są emeryci, najmłodsi są studenci, o dziwo pokójówki są stosunkowo stare.
```
