```python
np.random.seed(213)
```

```python
data = pd.read_csv("abalone_dataset.csv")
```

```python
corr = data.corr()
```

```python
ax = sns.heatmap(corr, 
            xticklabels=corr.columns,
            yticklabels=corr.columns, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('Korelacja zmiennych')
plt.show()
```

```python
sns.pairplot(data.iloc[:,[1, 2, 3, 8]], size=2)
plt.tight_layout()
```

```python
sexes = data[data.sex.isin(["M", "F", "I"])]
sns.violinplot(sexes.sex, sexes.height)
```

```python
plot_dens = sns.distplot(data.rings, hist = True, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 3})
plot_dens.set_title('RozkÅ‚ad zmiennej rings')
```
