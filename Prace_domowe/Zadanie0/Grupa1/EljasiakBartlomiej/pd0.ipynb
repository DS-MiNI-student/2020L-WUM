```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
abalone = pd.read_csv("abalone_dataset.csv")
```

```python
abalone.head()
```

```python
### Wszystkie dane są poprawne
```

```python
abalone.info()
```

```python
plot_dens = sns.distplot(abalone.height,hist=True, kde= True, 
                        kde_kws = {'shade': True})
plot_dens.set_title('Rozkład wysokości')
```

```python
sns.pairplot(abalone.iloc[:,[1,3,4]],height=2)
plt.tight_layout()
```

```python
corr= abalone.iloc[:,np.r_[0:8]].corr()
ax = sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns, annot=True)
bottom,top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('Korelacja zmiennych')
plt.show()
```

```python
sns.lmplot(x='diameter', y='whole_weight', data = abalone)
```

```python
sns.lmplot(x='shell_weight', y='shucked_weight', data = abalone)
```
