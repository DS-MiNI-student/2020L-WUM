```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/samba/ficp/Downloads/abalone_dataset.csv")
```

```python
df.info()
# brak braków obserwacji
```

```python
df.describe()
```

```python
plot_dens_length = sns.distplot(df.length, hist = True, kde = True)
plot_dens_length.set_title("Rozkład zmiennej length")
```

```python
plot_dens_height = sns.distplot(df.height, hist = True, kde = True)
plot_dens_height.set_title("Rozkład zmiennej height")
```

```python
plot_dens_diameter = sns.distplot(df.diameter, hist = True, kde = True)
plot_dens_diameter.set_title("Rozkład zmiennej diameter")
#Podobny rozklad do length, naturalnie rozmiary sa ze soba powiazane
```

```python
sns.pairplot(df.iloc[:, [1,2,3]])
plt.tight_layout()
# podobne obserwacje co na podstawie rozkladow, dotyczy diameter i length
```

```python
corr = df.corr()
ax=sns.heatmap(corr, 
            xticklabels=corr.columns,
            yticklabels=corr.columns, annot=True)

bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('Korelacja zmiennych')
plt.show()
```

```python
sex_weight = df.groupby('sex').whole_weight.mean()
sex_weight.plot(kind = 'bar', title = 'Średnia waga względem płci')
```

```python
sex_height = df.groupby('sex').height.mean()
sex_height.plot(kind = 'bar', title = 'Średnia wysokość względem płci')
```

```python
sex_rings = df.groupby('sex').rings.mean()
sex_rings.plot(kind = 'bar', title = 'Średni wiek względem płci')
```
