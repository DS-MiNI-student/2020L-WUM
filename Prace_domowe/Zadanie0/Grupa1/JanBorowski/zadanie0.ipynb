```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Wczytywanie danych 
df=pd.read_csv('/home/samba/borowskij/Downloads/abalone_dataset.csv', sep=',')
```

```python
plot_dens=sns.distplot(df.height, hist = True, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 3})
plot_dens.set_title('Rozkład zmiennej diameter')
# Rozkład normalnu ale uciety dla wartowści mniejsze niz zero z oczywistych powodów
```

```python

corr=df.iloc[:,np.r_[0:9]].corr()

ax=sns.heatmap(corr, 
            xticklabels=corr.columns,
            yticklabels=corr.columns, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('Korelacja zmiennych')
plt.show()
# Wszystkie zmienne od length do shell_weight sa bardzo mocno skorelowane 
```

```python
Male_Famale=df[df.sex.isin(['M', 'F','I'])]
sns.violinplot(Male_Famale.sex, Male_Famale.diameter)
# Uzywam jednej ze zmiennych ze wzgledu naich wysoka korelacje
```
