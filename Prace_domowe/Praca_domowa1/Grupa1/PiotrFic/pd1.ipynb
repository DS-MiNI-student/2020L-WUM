# Rozkłady zmiennych ciągłych

```python
# Funkcja do rysowania histogramu oraz boxplotu
# zmiennych ciągłych
def plot_var(col, variable):
    plot = sns.distplot(col, hist = True, kde = True)
    plot.set_title("Rozkład zmiennej " + variable)
    
    fig, ax = plt.subplots()
    ax.set_title("Zmienna " + variable)
    ax.boxplot(col, vert = False)
```

## Długość kredytu

```python
plot_var(df.duration, "długość kredytu")
# Rozkład asymetryczny przesunięty w kierunku długości kredytu
# krótszej niż ok. 20 lat, jednak zauważalna jest popularność 
# kilku konkretnych długości kredytów również w grupie 
# dłuższych niż 20 lat
```

```python
# Popularne długości kredytów dłuższych niż 20 lat
long_duration = df[df.duration>20]
popular_long_durations = long_duration["duration"].value_counts("duration").head(5)
popular_long_durations.plot(kind = 'bar', title = 'Popularne długości kredytów dłuższych niż 20 lat')
```

## Wartość kredytu

```python
plot_var(df.credit_amount, "wartość kredytu")
# Rozkład asymetryczny, ewidentnie dominują kredyty na
# kwoty do 5000 DM (ok. 11 000 zł), kredyty na wyższe kwoty tworzą istotny ogon rozkładu
```

## Wiek kredytobiorcy

```python
plot_var(df.age, "wiek kredytobiorcy")
# Ponownie rozkład asymetryczn analogiczny do poprzednich,
# im starsza grupa wiekowa tym mniej kredytobiorców
```

## Korelacje

```python
#Zmienna customer_type jest zmienną kategoryczną zapisaną liczbowo
#przez co nie była usuwana automatycznie przez corr()
#dla pozostałych zmiennych możemy wyznaczyć korelację
corr = df.drop("customer_type", axis = 1).corr()

ax=sns.heatmap(corr, xticklabels=corr.columns,
               yticklabels=corr.columns, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('Korelacje zmiennych')
plt.show()
# Jedyna znacząca korelacja pomiędzy zmiennymi dotyczy wyżej analizowanego przypadku
# długość kredytu - kwota kredytu
```

# Zmienne zgrupowane

## Stan konta (checking_account_status)

```python
sns.countplot(x = 'checking_account_status', data = df,
             order = ['A11','A12','A13','A14'])
# A11: <0, A12: 0<x<200, A13: >200, A14: no account
# Spora część kredytobiorców nie posiada konta w tym banku
# Niewielka część klientów posiada na koncie środki większe niż
# 200DM
```

## Oszczędności

```python
sns.countplot(x = 'savings', data = df,
             order = ['A61','A62','A63','A64', 'A65'])
# A61: <100, A62: 100<x<500, A63: 500<x<1000, A64: >10000, A65: unknown
# Wśród kredytobiorców dominują niskie oszczędności
```

Oszczędności vs dane kredytowe

```python
mean_amount_savings = df.groupby('savings').credit_amount.mean()
mean_amount_savings.plot(kind = 'bar', 
                         title = 'Kwota kredytu a oszczędności')
# Ciężko z pełną pewnością opisać wynik dla osób, które nie podały
# kwoty swoich osczędności, średnia kwota kredytu jest najwyższa w tej grupie
# Dla deklarujących kwotę oszczędności, średnia wartość kredytu
# jest wyższa w grupach o niższych oszczędnościach
```

```python
mean_duration_savings = df.groupby('savings').duration.mean()
mean_duration_savings.plot(kind = 'bar', 
                           title = 'Długość kredytu a oszczędności')
#W przypadku średniej długości kredytu różnice są mniejsze względem 
#grupy nie podającej oszczędności, pozostałe tendencje są analogiczne
#co daje istotne wnioski o związku kwoty i długości kredytu 
#a oszczędnościami kredytobiorcy
```

## Długość obecnego zatrudnienia

```python
sns.countplot(x = 'present_employment', data = df,
             order = ['A71','A72','A73','A74', 'A75'])
# A72: <1 year, A73: 1<x<4, A74: 4<x<7, A75: >7, A71: unemployed
# Kredyty rzadko udzielane są osobą bezrobotnym,
# najliczniejsze grupy to zatrudnienie 1-4 lat i ponad 7
```

Zatrudnienie vs dane kredytowe

```python
mean_amount_emp = df.groupby('present_employment').credit_amount.mean()
mean_amount_emp.plot(kind = 'bar', 
                         title = 'Kwota kredytu a zatrudnienie')
# Niespodziewanie osoby bezrobotne otrzymują kredyty na średnio 
# największą kwotę co może nie być najbezpieczniejszym 
# posunięciem ze strony banku
```

```python
mean_duration_emp = df.groupby('present_employment').duration.mean()
mean_duration_emp.plot(kind = 'bar', 
                           title = 'Długość kredytu a zatrudnienie')
# Osoby o stabilnym zatrudnieniu (dłuższy staż), pozwalają sobie
# na średnio najdłuższe kredyty
```

# Zmienne kategoryczne

```python
categorical = ['credit_history', 'purpose', 'personal', 'housing']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
axes = axes.flatten()
for idx, ax in enumerate(axes):
    plt.sca(ax)
    ax = sns.countplot(x = categorical[idx], data=df,
                      order = df[categorical[idx]].value_counts().index)
sns.despine()
plt.tight_layout()
plt.show()

# Credit history:
# - dominujące dwie kategorie to osoby spłacające raty na bieżąco
#   i konto krytyczne / kredyty w innym banku
# Personal: 
# - najliczniejsza kategoria to mężczyźni-single, potem 
#   kobiety-rozwiedzione/zamężne
# Housing:
# - najliczniejsza kategoria to własna posiadłość, 
#   następnie wynajmowana
```
